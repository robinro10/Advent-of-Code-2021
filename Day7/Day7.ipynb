# Partie 1

```python
import numpy as np

data = np.sort(np.genfromtxt("input7.txt", delimiter=",", dtype="int"))
data
```

```python
med = np.median(data)

print(med)

sum(abs(data - med).astype(int))
```

# Partie 2

```python
mean = int(np.mean(data))
mean
```

```python
updatedData = np.array([sum(range(abs(data[i] - mean) + 1)) for i in range(len(data))])
updatedData2 = np.array([sum(range(abs(data[i] - (mean + 1)) + 1)) for i in range(len(data))])
```

```python
min(sum(updatedData), sum(updatedData2))
```
