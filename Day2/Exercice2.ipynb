```python
import pandas as pd

data = pd.read_csv("input2.txt", delimiter = " ", header = None)
data
```

```python
forward = data[data[0] == "forward"][1].sum()
forward
```

```python
down = data[data[0] == "down"][1].sum()
down
```

```python
up = data[data[0] == "up"][1].sum()
up
```

```python
(down - up)*forward
```

# Partie 2

```python
data["aim"] = (data[data[0] == "down"][1].reindex(data.index, fill_value=0) - data[data[0] == "up"][1].reindex(data.index, fill_value=0)).cumsum()
data
```

```python
data["x"] = data[data[0] == "forward"][1].cumsum()
data = data.ffill()
data
```

```python
data["depth"] = data[data[0] == "forward"][1] * data[data[0] == "forward"]["aim"]
data["depth"] = data["depth"].fillna(0).cumsum()
data
```

```python
data["depth"].iloc[-1] * data["x"].iloc[-1]
```
